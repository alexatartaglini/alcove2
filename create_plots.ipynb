```python
import glob
import pandas as pd
import altair as alt
import numpy as np
import csv
```

```python
human_results_df = pd.read_csv('csv/human/human.csv',index_col=0, header=0,engine='python')
base_human = alt.Chart(human_results_df).mark_line(clip=True).encode(
    x=alt.X('Block', scale=alt.Scale(domain=(0, 16))),
    y=alt.Y('Probability Correct', scale=alt.Scale(domain=(0, 1))),
    color='Type:N'
).properties(title='Human Results', width=480, height=360)

base_human
base_human.save('img/human.svg')
```

```python
alcove_ab_group_cols = ['Model', 'Loss Type', 'Image Set', 'LR-Attention', 'LR-Association', 'c', 'phi']
df = pd.read_csv('csv/alcove_ab/summary.csv')
df_configs = df.drop_duplicates(subset=alcove_ab_group_cols)

configs_df = df.groupby(alcove_ab_group_cols)
for config_df_name, config_df in configs_df:
    # convert elements to str to get filename
    config_df_name = [str(x) for x in config_df_name]
    config_df_name = '-'.join(config_df_name)
    print(config_df_name)

    model_plot = alt.Chart(config_df).mark_line().encode(
        x=alt.X('Epoch', scale=alt.Scale(domain=(0, 128)),
               axis=alt.Axis(values=[0, 16, 32, 48, 64, 80, 96, 112, 128])),
        y=alt.Y('Probability Correct', scale=alt.Scale(domain=(0, 1))),
        color=alt.Color('Type:N')
    ).properties(title='ALCOVE - Abstract Stimuli', width=480, height=360)
    
    model_filename = f'img/{config_df_name}.svg'
    model_plot.save(model_filename)
```

```python
alcove_im_group_cols = ['Model', 'Net', 'Loss Type', 'Image Set', 'LR-Attention', 'LR-Association', 'c', 'phi']
df = pd.read_csv('csv/alcove_im/summary.csv')
configs_df = df.groupby(alcove_im_group_cols)
for idx, (config_df_name, config_df) in enumerate(configs_df):
    # convert elements to str to get filename
    config_df_name = [str(x) for x in config_df_name]
    config_df_name = '-'.join(config_df_name)
    print(idx, config_df_name)

    model_plot = alt.Chart(config_df).mark_line().encode(
        x=alt.X('Epoch', scale=alt.Scale(domain=(0, 128)),
               axis=alt.Axis(values=[0, 16, 32, 48, 64, 80, 96, 112, 128])),
        y=alt.Y('Probability Correct', scale=alt.Scale(domain=(0, 1))),
        color=alt.Color('Type:N')
    ).properties(title='ALCOVE - Image Stimuli', width=480, height=360)
    
    model_filename = f'img/{config_df_name}.svg'
    model_plot.save(model_filename)
```
